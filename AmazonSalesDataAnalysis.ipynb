```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Amazon Sales data.csv")
df.head()
```

```python
df['Order ID'].duplicated()
```

```python
df.drop_duplicates('Order ID')
```

```python
df.describe()
```

```python
sns.histplot(x='Total Revenue', data=df, hue='Sales Channel', element='poly')
```

```python
sns.barplot(y='Region',x='Units Sold',data=df,palette='CMRmap')
```

```python
sns.histplot(x='Total Profit', data=df, hue='Item Type')
```

```python
sns.boxplot(x='Order Priority', y='Total Profit', data=df, color='pink', linewidth=0.5)
```

```python
sns.barplot(x='Unit Cost', y='Item Type', data=df, palette='CMRmap')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Order Year'] = df['Order Date'].dt.year
```

```python
sns.histplot(y='Order Date',hue='Region',data=df)
```

```python
sns.lineplot(x='Order Date', y='Total Profit', data=df, marker='o', color='black')
sns.set(rc={'figure.figsize':(15,10)})
plt.title('Year on Year Growth',fontsize=30)
plt.xlabel('Order Year',fontsize=25)
plt.ylabel('Total Profit',fontsize=25)
```

```python
sns.barplot(x='Region', y='Total Profit', data=df, palette='cividis', ci=None)
```
